###Classify Movies Based on Ratings [Excellent, Good, and Average]

```python
def rating_cat(rating):
    if rating >=7 :
        return 'Excellent'
    elif rating >= 6.0:
        return 'Good'
    else:
        return 'Average'

data['Rating Category'] = data['Rating'].apply(rating_cat)
sns.boxplot(data=data, x="Rating Category", y="Revenue (Millions)")
```
